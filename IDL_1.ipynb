```python
import pip


class Trabajador:
    def __init__(self, nombre, estado="Activo"):
        self.nombre = nombre
        self.estado = estado
    
    def get_resumen(self):
        return "Trabajador genérico"

    def get_jefe_inmediato(self):
        return "No tiene jefe inmediato"

    def get_estado(self):
        return self.estado

    # Getters y Setters generales
    def get_nombre(self):
        return self.nombre
    
    def set_nombre(self, nombre):
        self.nombre = nombre
    
    def set_estado(self, estado):
        self.estado = estado


class Gerente(Trabajador):
    def __init__(self, nombre):
        super().__init__(nombre)
    
    def get_resumen(self):
        return "Gerente"
    

class JefeDeArea(Trabajador):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.asistentes = []
        self.tecnicos = []

    def agregar_asistente(self, asistente):
        if len(self.asistentes) < 2:
            self.asistentes.append(asistente)
        else:
            print(f"No se pueden agregar más asistentes a {self.nombre}")

    def agregar_tecnico(self, tecnico):
        if len(self.tecnicos) < 5:
            self.tecnicos.append(tecnico)
        else:
            print(f"No se pueden agregar más técnicos a {self.nombre}")

    def get_resumen(self):
        return "Jefe de Área"
    
    def get_jefe_inmediato(self):
        return "Gerente"


class Asistente(Trabajador):
    def __init__(self, nombre):
        super().__init__(nombre)
    
    def get_resumen(self):
        return "Asistente"
    
    def get_jefe_inmediato(self):
        return "Jefe de Área"


class Tecnico(Trabajador):
    def __init__(self, nombre, experiencia):
        super().__init__(nombre)
        self.experiencia = experiencia
    
    def get_resumen(self):
        return f"Técnico con {self.experiencia} años de experiencia"
    
    def get_jefe_inmediato(self):
        return "Jefe de Área"
    
    def set_experiencia(self, experiencia):
        self.experiencia = experiencia

# Crear trabajadores
gerente = Gerente("Juan Pérez")

jefe_marketing = JefeDeArea("Ana Gómez")
jefe_sistemas = JefeDeArea("Carlos Ramírez")
jefe_produccion = JefeDeArea("Luisa Torres")
jefe_logistica = JefeDeArea("María Rodríguez")

asistente_marketing1 = Asistente("Pedro Sánchez")
asistente_marketing2 = Asistente("Laura Martínez")
asistente_sistemas1 = Asistente("Roberto Díaz")
asistente_produccion1 = Asistente("Marta González")
asistente_logistica1 = Asistente("Javier Fernández")

tecnico1 = Tecnico("Sofía López", 3)
tecnico2 = Tecnico("Daniel Ruiz", 5)
tecnico3 = Tecnico("Lucía Vargas", 2)
tecnico4 = Tecnico("Diego Castro", 4)
tecnico5 = Tecnico("Elena Jiménez", 1)

# Asignar asistentes y técnicos a los jefes de área
jefe_marketing.agregar_asistente(asistente_marketing1)
jefe_marketing.agregar_asistente(asistente_marketing2)
jefe_sistemas.agregar_asistente(asistente_sistemas1)
jefe_produccion.agregar_asistente(asistente_produccion1)
jefe_logistica.agregar_asistente(asistente_logistica1)

jefe_marketing.agregar_tecnico(tecnico1)
jefe_sistemas.agregar_tecnico(tecnico2)
jefe_produccion.agregar_tecnico(tecnico3)
jefe_produccion.agregar_tecnico(tecnico4)
jefe_logistica.agregar_tecnico(tecnico5)

# Crear el array de objetos
trabajadores = [
    gerente,
    jefe_marketing,
    jefe_sistemas,
    jefe_produccion,
    jefe_logistica,
    asistente_marketing1,
    asistente_marketing2,
    asistente_sistemas1,
    asistente_produccion1,
    asistente_logistica1,
    tecnico1,
    tecnico2,
    tecnico3,
    tecnico4,
    tecnico5
    ]
```
